```python
import os
import shutil
fold=0
data_for='train'


# Paths
image_names_dir = f'/home/patel_zeel/kilns_neurips24/crossval/lucknow_sarath_grid_obb_v3/{fold}/{data_for}/labels'
source_dir = f'/home/patel_zeel/kilns_neurips24/crossval/lucknow_sarath_grid_obb_v3/{fold}/{data_for}/labels'
target_dir = f'/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/super_res_exp/Time-Series-Super-Resolution/crossval_sr/lucknow_sarath_grid_obb_v3/{fold}/{data_for}/labels'

# Create target directory if it doesn't exist
os.makedirs(target_dir, exist_ok=True)

# Get the list of image filenames to be copied
image_filenames = os.listdir(image_names_dir)

# Copy the images
for image_filename in image_filenames:
    source_path = os.path.join(source_dir, image_filename)
    target_path = os.path.join(target_dir, image_filename)
    
    # Check if the source image exists
    if os.path.exists(source_path):
        shutil.copy(source_path, target_path)
        print(f'Copied: {image_filename}')
    else:
        print(f'Source image not found: {image_filename}')

print('Image copying completed.')
```
